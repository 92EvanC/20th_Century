```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
```

```python
# Setup the ChromeDriver using webdriver-manager
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
```

```python
# Open the webpage (replace with your actual URL)
driver.get('https://en.wikipedia.org/wiki/20th_century')
```

```python
# You can now interact with the page using the driver object
# For example, print the page title
print(driver.title)
```

```python
# Get the page source after loading the content
html = driver.page_source
```

```python
# Parse the page with BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Find the elements you want to scrape (adjust this based on the page structure)
events = soup.find_all('li')  # Assuming the list items contain events
```

```python
# Check what we're getting
for event in events:
    print(event.text)  # Print each event
```

```python
# Save the content into a .txt file
with open('key_events_20th_century.txt', 'w') as f:
    for event in events:
        f.write(event.text + '\n')
```

```python
# Close the driver
driver.quit
```
